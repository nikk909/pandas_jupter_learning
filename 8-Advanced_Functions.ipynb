# 1.shift移动

```python
import pandas as pd
coffee = pd.read_csv("./warmup-data/coffee_updated.csv")
```

```python
#因为有两行所以移2下，然后可以比较上周的收入和这周的收入
coffee["yesterday_revenue"] = coffee["revenue"].shift(2)
```

```python
coffee.head()
```

# 2.rank排序

```python
bios = pd.read_csv("./data/bios.csv")
```

```python
bios.head()
```

```python
bios["height_rank"] = bios["height_cm"].rank()
```

```python
bios.head()
```

```python
bios.sort_values(["height_rank"],ascending = False)
```

# 3.cumsum累加/cumulative

```python
coffee.head()
```

```python
coffee["cumsum_revenue"] = coffee["revenue"].cumsum()
```

```python
coffee.head()
```

# 3. rolling滚动窗口

```python
latte = coffee[coffee["Coffee Type"] == "Latte"].copy()
latte
```

```python
latte["3day Units Sold"] = latte["Units Sold"].rolling(3).sum()
```

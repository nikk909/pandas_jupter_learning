# 4. Filtering data

## 4.1 conditional request条件查询

```python
bios.head()
```

```python
bios.tail()
```

```python
bios.info()
```

```python
bios.loc[bios["height_cm"] > 215]
```

```python
bios.loc[bios["height_cm"] > 215,["name","height_cm"]]
```

```python
#shorthand syntax简写的语法
bios[bios["height_cm"] > 215]
```

```python
#跟矩阵用法或是二维数组索引一样，但是要注意括号uu
bios[bios["height_cm"] > 215][["name","height_cm"]]
```

```python
#注意两个表达式需要有括号
bios[(bios["height_cm"] > 215) & (bios["born_country"] == "USA")]
```

```python
bios[bios["name"].str.contains("keith")]
```

```python
bios[bios["name"].str.contains("Keith")]
```

```python
#case = false的意思是无视大小写，默认为true也就是大小写敏感
bios[bios["name"].str.contains("keith", case=False)]
```

```python
#引号里可以使用正则表达式regular expression
#正则表达式又回到了那些$r之类的可以表示连续或是只包含一个的强大的功能
bios[bios["name"].str.contains("keith|Patrick", case=False)]
```

```python
#regex = False的意思是把引号里的内容当成普通的字符串，而不是正则表达式,因此找不到完全匹配的
bios[bios["name"].str.contains("keith|Patrick", case=False,regex = False)]
```

```python
#记得矩阵匹配是包含在括号里的，而不是单另直接写括号就行
bios[bios["born_country"].isin(["USA","UK"])]
```

```python
#是startswith而不是startwith
#可以在表达式里直接用&合并
bios[bios["born_country"].isin(["USA","UK"]) & bios["name"].str.startswith("Keith")]
```

```python
#另一种写法，query，记得引号里是and,还有字符串需要单引号包含双引号或是双引号包含单引号
bios.query("born_country == 'USA' and born_city == 'Seattle'")
```

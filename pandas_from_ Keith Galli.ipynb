3. accessing data with pandas

3.1 看头部/尾部/随机数

```python
coffee.head()
```

```python
#看全部的数据
coffee
#其他的语法 print(coffee) 或是 display(coffee) 都可以显示全部数据,在当前的环境里也可以显示
```

```python
#如果觉得数据开头结尾不是很重要并且数据量很大或是很分散，可以用sample随机抽取一些行来看看
# 括号里是采样指定的行数
coffee.sample(10)
#重复执行结果不一样，因为是随机抽取的
```

```python
coffee.sample(10,random_state = 1)
#random_state参数可以让每次随机抽取的结果一样，参数值可以是任意整数，
# 或者是None（默认值），如果是None，每次抽取的结果都不一样
#如果是固定值，每次抽取的结果都一样
#底层逻辑是随机数种子
#可以叫做repeatable或是deterministic确定性的
```

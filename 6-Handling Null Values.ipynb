```python
import pandas as pd
```

```python
coffee = pd.read_csv("./warmup-data/coffee.csv")
```

```python
coffee.head()
```

```python
import numpy as np
coffee["price"] = np.where(
    coffee["Coffee Type"] == "Espresso",4.99,
    5.99
)
```

```python
coffee.head()
```

```python
coffee["revenue"] = coffee["price"] * coffee["Units Sold"]
```

```python
coffee.head()
```

```python
#设置两个coffee的price为nan
coffee.loc[[0,1],"Units Sold"] = np.nan
```

## 1.查看空值

```python
coffee.head()
```

```python
coffee.info()
```

```python
coffee.isna().sum()
```

## 2. 给nan赋值

```python
#直接赋值，注意这里也是有inpalce默认为false
coffee.fillna(10000)
```

```python
#赋予nan均值
coffee.fillna(coffee["Units Sold"].mean())
```

```python
#赋予nan插值,首尾连续的话没法赋予
coffee.fillna(coffee["Units Sold"].interpolate())
```

```python
coffee.fillna(15,inplace = True)
```

```python
coffee.loc[[2,3],"Units Sold"] = np.nan
```

```python
coffee.head()
```

```python
coffee["Units Sold"].interpolate()
```

```python
coffee["Units Sold"] = coffee["Units Sold"].interpolate()
```

```python
coffee.head()
```

```python
coffee.loc[[1,2,3],["Units Sold"]] = np.nan
```

```python
coffee.loc[[1,2],"price"] = np.nan
```

```python
coffee
```

# 3.Drop nan 删掉空值

```python
coffee.dropna()
```

```python
#只去掉subet一列的nan
coffee.dropna(subset = ["price"])
```

```python
#drop函数也需要写inplace = True 或是赋值给原本的变量
#获取为nan的行
coffee[coffee["Units Sold"].isna()]
```

```python
#获取不为nan的行
coffee[coffee["Units Sold"].notna()]
```

```python
coffee.to_csv("./warmup-data/coffee_updated.csv",index = False)
```

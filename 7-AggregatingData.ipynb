```python
coffee = pd.read_csv("./warmup-data/coffee.csv")
```

```python
coffee.head()
```

```python
coffee.groupby("Coffee Type")["Units Sold"].sum()
```

```python
coffee.groupby("Coffee Type")["Units Sold"].mean()
```

```python
coffee["Price"] = coffee["Coffee Type"].apply(lambda x: 4.99 if x == "Espresso" else 5.99)
```

```python
#groupby算出中间过程，agg传入字典，指定不同聚合函数的结果是求啥，平均还是求和还是啥
coffee.groupby("Coffee Type").agg({
    "Units Sold":"sum",
    "Price":"mean"})
```

```python
coffee.groupby(["Coffee Type","Day"]).agg({
    "Units Sold":"sum",
    "Price":"mean"})
```

# 2.pivot table枢纽，透视表

```python
coffee["revenue"] = coffee["Units Sold"] * coffee["Price"]
```

```python
pivot = coffee.pivot(columns = "Coffee Type",index = "Day",values = "revenue")
```

```python
pivot
```

```python
#有新的方式访问
#转成行列+表格里的值的写法
pivot.loc["Monday","Latte"]
```

```python
#列加和不包括index
pivot.sum()
```

```python
#行加和
pivot.sum(axis = 1)
```

```python
bios["born_date"] = pd.to_datetime(bios["born_date"])
```

```python
#dt.year访问datetime类型的年份
bios.groupby(bios["born_date"].dt.year)["name"].count()
```

```python
#先给一个序列index然后递减排序
bios.groupby(bios["born_date"].dt.year)["name"].count().reset_index().sort_values("name",ascending = False)
```

```python
bios["born_year"] = bios["born_date"].dt.year
bios["born_month"] = bios["born_date"].dt.month

bios.groupby(["born_year", "born_month"])["name"].count().reset_index().sort_values("name", ascending=False)
```

```python
#也可以写成
bios.groupby([bios["born_year"], bios["born_month"]])["name"].count().reset_index().sort_values("name", ascending=False)
```

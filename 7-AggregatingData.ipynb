```python
coffee = pd.read_csv("./warmup-data/coffee.csv")
```

```python
coffee.head()
```

```python
coffee.groupby("Coffee Type")["Units Sold"].sum()
```

```python
coffee.groupby("Coffee Type")["Units Sold"].mean()
```

```python
coffee["Price"] = coffee["Coffee Type"].apply(lambda x: 4.99 if x == "Espresso" else 5.99)
```

```python
#groupby算出中间过程，agg传入字典，指定不同聚合函数的结果是求啥，平均还是求和还是啥
coffee.groupby("Coffee Type").agg({
    "Units Sold":"sum",
    "Price":"mean"})
```

```python
coffee.groupby("Coffee Type","Day").agg({
    "Units Sold":"sum",
    "Price":"mean"})
```
